```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
```

CREATE 2 ARRAYS AND SET THE LABELS & LIMITS
eg:x: starting value 0
      last value 50
      number of elements 100

```python
x = np.linspace(0,50,100)
y = x * np.linspace(100,150,100)

plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.xlim(0,45)
plt.ylim(1200, 7000)
```

CREATING SUBPLOTS AND STACKING THEM

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')
for ax in axs.flat:
    ax.label_outer()
```

ADDING A LEGEND TO A PLOT

```python
a = [10,20,30,40,50]
y1 = [1, 2, 3, 4, 5]
y2 = [1, 4, 9, 16, 25]
plt.plot(a, y1,color='purple', label ='y1')
plt.plot(a, y2,color='blue',label ='y2')
plt.legend(loc="lower right")
```




IRIS DATASET SCATTER PLOT

```python
from sklearn import datasets
# Load Iris dataset
iris = datasets.load_iris()

# Extract the features and target variable
X = iris.data
y = iris.target

# Plot the scatter plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')

# Set labels for the scatter plot
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset: Sepal Length vs Sepal Width')

# Show the plot
plt.show()
```

IRIS DATASET BUBBLE PLOT

```python
petal_length = X[:, 2]

# Plot the bubble plot
plt.scatter(X[:, 0], X[:, 1], s=petal_length*50, c=y, cmap='viridis')
```

GROUPED BAR PLOT

```python
m=[9,7,8,4,2]
width = 0.40


plt.bar(x-0.2, m, width,color="purple")
plt.bar(x+0.2, n, width,color="pink")
```

  STACKED BAR PLOT

```python
l = [4, 6, 2, 8, 9]
plt.bar(x, n,color="pink")
plt.bar(x, m, bottom=n,color="aqua")
plt.bar(x, l, bottom=np.add(n, m), color='blue')

plt.show()
```

CONTOUR PLOT

```python
X,N=np.meshgrid(x,n)
Z = np.sin(np.sqrt(X**2 + N**2))
plt.contour(X,N,Z)
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Contour Plot')

plt.show()
```
